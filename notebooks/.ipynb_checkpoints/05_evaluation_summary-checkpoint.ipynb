# 05 — Evaluation Summary

**Notebook:** `notebooks/05_evaluation_summary.ipynb`  
**Purpose:** Load `best_model.pkl`, verify metrics against `metrics.json`, generate evaluation plots, and confirm the pipeline is inference-ready for `app.py`.  
**Target variable:** `log_travel_time` (back-transformed via `np.expm1()` → reported in minutes)

````python
import pickle
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
from pathlib import Path
from sklearn.pipeline import Pipeline

# ── Paths ──────────────────────────────────────────────────────────────
MODEL_PATH      = Path("../models/best_model.pkl")
METRICS_PATH    = Path("../models/metrics.json")
FI_PATH         = Path("../models/feature_importance.csv")
X_TEST_PATH     = Path("../data/processed/splits/X_test.csv")
Y_TEST_PATH     = Path("../data/processed/splits/y_test.csv")

USE_LOG_TRANSFORM = True   # must match metrics.json → "log_transform": true

print("Imports OK")
````

````python
# ── Load model ──────────────────────────────────────────────────────────
with open(MODEL_PATH, "rb") as f:
    model = pickle.load(f)

assert isinstance(model, Pipeline), "Loaded object is not an sklearn Pipeline!"
print(f"✅ Model loaded from: {MODEL_PATH}")
print(f"   Pipeline steps: {[s[0] for s in model.steps]}")

# ── Load saved metrics ──────────────────────────────────────────────────
with open(METRICS_PATH, "r") as f:
    saved_metrics = json.load(f)

print(f"\n✅ metrics.json loaded")
print(f"   Model: {saved_metrics['model_name']}")
print(f"   Train rows: {saved_metrics['training_rows']:,}  |  Test rows: {saved_metrics['test_rows']:,}")

# ── Load feature importance ─────────────────────────────────────────────
fi_df = pd.read_csv(FI_PATH)
print(f"\n✅ feature_importance.csv loaded — {len(fi_df)} features")
fi_df.head()
````

````python
X_test = pd.read_csv(X_TEST_PATH)

# Force 1D Series regardless of how the CSV was saved
y_test_raw = pd.read_csv(Y_TEST_PATH)
print(f"y_test raw shape before squeeze : {y_test_raw.shape}")
print(f"y_test columns : {list(y_test_raw.columns)}")

# Drop any unnamed index columns, keep only the target column
target_col = [c for c in y_test_raw.columns if 'log_travel_time' in c]
if target_col:
    y_test = y_test_raw[target_col[0]].reset_index(drop=True)
else:
    # Fallback: drop index-like columns and take the remaining single column
    y_test_raw = y_test_raw.loc[:, ~y_test_raw.columns.str.contains('^Unnamed')]
    y_test = y_test_raw.iloc[:, 0].reset_index(drop=True)

print(f"\nX_test shape : {X_test.shape}")
print(f"y_test shape : {y_test.shape}   ← must be 1D")
print(f"y_test dtype : {y_test.dtype}")
print(f"y_test sample:\n{y_test.head()}")
````

````python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Confirm y_test is 1D before proceeding
assert y_test.ndim == 1, f"y_test must be 1D — got shape {y_test.shape}"

# Raw predictions (log-space)
y_pred_log = model.predict(X_test)

# Back-transform
if USE_LOG_TRANSFORM:
    y_pred_raw = np.expm1(y_pred_log)
    y_true_raw = np.expm1(y_test.to_numpy())   # .to_numpy() guarantees flat 1D array
else:
    y_pred_raw = y_pred_log
    y_true_raw = y_test.to_numpy()

# ── Diagnostics ─────────────────────────────────────────────────────────
print(f"y_true_raw shape : {y_true_raw.shape}")
print(f"y_pred_raw shape : {y_pred_raw.shape}")
print(f"y_true — inf count : {np.sum(np.isinf(y_true_raw))}")
print(f"y_true — nan count : {np.sum(np.isnan(y_true_raw))}")
print(f"y_pred — inf count : {np.sum(np.isinf(y_pred_raw))}")
print(f"y_pred — nan count : {np.sum(np.isnan(y_pred_raw))}")
print(f"\ny_test (log-space) max : {y_test.max():.4f}")
print(f"y_test (log-space) min : {y_test.min():.4f}")

# ── Drop rows where either value is inf/nan/non-positive ────────────────
finite_mask = (
    np.isfinite(y_true_raw) &
    np.isfinite(y_pred_raw) &
    (y_true_raw > 0) &
    (y_pred_raw > 0)
)
dropped = (~finite_mask).sum()
print(f"\nRows dropped (inf/nan/non-positive) : {dropped:,} "
      f"({dropped / len(finite_mask) * 100:.2f}%)")

y_true = y_true_raw[finite_mask]
y_pred = y_pred_raw[finite_mask]

# ── Compute metrics ─────────────────────────────────────────────────────
mae  = mean_absolute_error(y_true, y_pred)
mse  = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
r2   = r2_score(y_true, y_pred)
mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100

live_metrics = {"MAE": mae, "MSE": mse, "RMSE": rmse, "R2": r2, "MAPE": mape}

print("\n📊 Live metrics (re-computed from best_model.pkl — finite rows only):")
for k, v in live_metrics.items():
    print(f"   {k:>6}: {v:.4f}")
````

````python
TOLERANCE = 0.01   # allow 1% relative difference

print("═" * 60)
print(f"{'Metric':<8}  {'Saved':>12}  {'Live':>12}  {'Δ%':>8}  Status")
print("═" * 60)

all_ok = True
for key in ["MAE", "MSE", "RMSE", "R2", "MAPE"]:
    saved_val = saved_metrics[key]
    live_val  = live_metrics[key]
    # relative difference — guard against zero denominator
    denom = abs(saved_val) if abs(saved_val) > 1e-9 else 1e-9
    rel_diff = abs(live_val - saved_val) / denom
    status = "✅" if rel_diff <= TOLERANCE else "⚠️  MISMATCH"
    if rel_diff > TOLERANCE:
        all_ok = False
    print(f"{key:<8}  {saved_val:>12.4f}  {live_val:>12.4f}  {rel_diff*100:>7.3f}%  {status}")

print("═" * 60)
if all_ok:
    print("\n✅ All metrics reproduced within tolerance — .pkl is valid.")
else:
    print("\n⚠️  One or more metrics differ. Check random_state or data path.")
````

## Metrics Summary

| Metric | Value | Interpretation |
|---|---|---|
| MAE | 7.83 min | Off by ~7.8 minutes on average — primary selection metric |
| RMSE | 35.72 min | High vs MAE → extreme outlier trips pulling squared error up |
| R² | 0.2054 | 20.5% of variance explained — low due to deliberate synthetic noise |
| MAPE | 37.78% | Percentage error; unstable on very short trips |
| Baseline MAE | 17.48 min | Mean predictor floor |
| **Improvement** | **55.18%** | **Target ≥ 30% → ✅ Met** |

> **R² not met (target > 0.70):** Expected behaviour — the dataset was engineered with ~8% Gaussian noise, 3–7% invalid FK rows, and injected outliers. These set a hard floor on achievable R².  
> **RMSE >> MAE:** Confirms a subset of extreme outlier trips dominates squared error.

````python
fig, ax = plt.subplots(figsize=(7, 6))

# Sample for readability (full 47K points are dense)
rng = np.random.default_rng(42)
idx = rng.choice(len(y_true), size=3000, replace=False)

ax.scatter(y_true[idx], y_pred[idx], alpha=0.25, s=8, color="steelblue", label="Predictions")
lim = max(y_true[idx].max(), y_pred[idx].max()) * 1.05
ax.plot([0, lim], [0, lim], "r--", lw=1.5, label="Perfect fit")

ax.set_xlabel("Actual travel time (min)", fontsize=12)
ax.set_ylabel("Predicted travel time (min)", fontsize=12)
ax.set_title("Actual vs Predicted Travel Time — Random Forest", fontsize=13, fontweight="bold")
ax.legend()
ax.set_xlim(0, lim); ax.set_ylim(0, lim)
plt.tight_layout()
plt.show()

print("💡 Insight: Points cluster along the diagonal for short/medium trips. "
      "Long-tail outlier trips (top-right scatter) pull RMSE up while leaving MAE relatively stable.")
````

````python
residuals = y_true - y_pred

fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(y_pred[idx], residuals[idx], alpha=0.2, s=8, color="darkorange")
ax.axhline(0, color="black", lw=1.2, linestyle="--")

ax.set_xlabel("Predicted travel time (min)", fontsize=12)
ax.set_ylabel("Residual  (actual − predicted, min)", fontsize=12)
ax.set_title("Residuals vs Predicted — Heteroscedasticity Check", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.show()

print("💡 Insight: Residual spread widens for longer predicted times — a "
      "classic heteroscedastic pattern. The model under-predicts extreme trips "
      "(positive residuals at the top), consistent with the injected outlier rows.")
````

````python
fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(residuals, bins=120, color="slateblue", edgecolor="none", alpha=0.85)
ax.axvline(0, color="red", lw=1.5, linestyle="--", label="Zero error")
ax.axvline(residuals.mean(), color="gold", lw=1.5, linestyle="-",
           label=f"Mean residual = {residuals.mean():.2f} min")

ax.set_xlabel("Residual (min)", fontsize=12)
ax.set_ylabel("Count", fontsize=12)
ax.set_title("Residual Distribution", fontsize=13, fontweight="bold")
ax.legend()
plt.tight_layout()
plt.show()

print(f"💡 Insight: Distribution centred near zero (mean = {residuals.mean():.2f} min) "
      "with heavy right tail from outlier trips. The model is not systematically biased "
      "but is affected by the synthetic noise floor.")
````

````python
fi_plot = fi_df.sort_values("importance", ascending=True).tail(15)

fig, ax = plt.subplots(figsize=(9, 6))
bars = ax.barh(fi_plot["feature"], fi_plot["importance"], color="teal", edgecolor="none")

ax.set_xlabel("Importance (mean decrease in impurity)", fontsize=12)
ax.set_title("Top-15 Feature Importances — Random Forest", fontsize=13, fontweight="bold")
ax.xaxis.set_major_formatter(mticker.PercentFormatter(xmax=1.0))

# Annotate each bar
for bar, val in zip(bars, fi_plot["importance"]):
    ax.text(bar.get_width() + 0.003, bar.get_y() + bar.get_height() / 2,
            f"{val:.3f}", va="center", fontsize=9)

plt.tight_layout()
plt.show()

print("💡 Insight: distance_km dominates at ~65% — matches the synthetic data "
      "generation logic where distance is the primary travel time driver. "
      "num_signals (9.4%) and avg_speed_kmph (3.7%) are the next most informative. "
      "Weather and binary flags contribute marginally but are non-zero.")
````
